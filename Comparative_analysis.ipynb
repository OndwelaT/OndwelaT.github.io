```python
# bar plot of comparing  Naive Bayes , F1 score ,precision, recall, accuracy with Few shot prompt for Gemini

import matplotlib.pyplot as plt
import numpy as np

# Replace these with your actual values
models = ['Naive Bayes','K-NN', 'SVM', 'Random Forest', 'Few-shot']
f1_scores = [0.78,0.67,0.68,0.65,0.72]
precisions = [0.78,0.65,0.68,0.67,0.74]
recalls = [0.79,0.69,0.74,0.71,0.72]
accuracies = [0.79,0.69,0.74,0.71, 0.72]

x = np.arange(len(models))  # the label locations
width = 0.2 # the width of the bars


fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - 3 * width / 2, f1_scores, width, label='F1 Score', color='skyblue')
rects2 = ax.bar(x - width / 2, precisions, width, label='Precision', color='salmon')
rects3 = ax.bar(x + width / 2, recalls, width, label='Recall', color='lightgreen')
rects4 = ax.bar(x + 3 * width / 2, accuracies, width, label='Accuracy', color='plum')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
fig.suptitle('Performance Comparison Traditional methods vs  Few-shot Prompt (Gemini)', fontsize=16, y=-0.1)
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=4)


fig.tight_layout()
plt.show()
```
