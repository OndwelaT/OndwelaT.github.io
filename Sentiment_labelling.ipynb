```python

import pandas as pd
from transformers import pipeline

# Load the sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Load the Excel file
df = pd.read_excel('/content/clean_text_data (1).xlsx')

# Display column names to verify the correct column
print("Columns in the dataframe:", df.columns)

# Assuming the text is in a column named 'clean_text'
# Drop duplicate rows based on the 'clean_text' column
df = df.drop_duplicates(subset='clean_text')

# Convert the 'clean_text' column to string type
df['clean_text'] = df['clean_text'].astype(str)

# Perform sentiment analysis and create a new column for sentiment labels
df['sentiment'] = df['clean_text'].apply(lambda x: classifier(x)[0]['label'])

# Save the results to a new Excel file
df.to_excel('/content/sentiment_labeled_data.xlsx', index=False)

# Display the first few rows of the dataframe to confirm the output
df.head()
# Perform sentiment analysis and create a new column for sentiment labels with a threshold for neutrality
df['sentiment'] = df['clean_text'].apply(lambda x: classifier(x)[0]['label'] if classifier(x)[0]['score'] > 0.6 else 'NEUTRAL')
```
